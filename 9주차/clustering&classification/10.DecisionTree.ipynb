```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
import graphviz
```

```python
dt_clf = DecisionTreeClassifier()
```

```python
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2)
dt_clf.fit(X_train, y_train)
```

```python
# export_graphviz()의 호출 결과로 out_file로 지정된 tree.dot파일을 생성함
export_graphviz(dt_clf, out_file="tree.dot", class_names=iris_data.target_names,
                feature_names=iris_data.feature_names, impurity=True, filled=True)

# 생성한 tree.dot파일을 읽어서 jupyter notebook 상에서 시각화
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
dot_data = export_graphviz(dt_clf, out_file=None, class_names=iris_data.target_names,
                           feature_names=iris_data.feature_names, impurity=True, filled=True, rounded=True)
graph = graphviz.Source(dot_data)
graph.render(filename='test', format='png')
```
