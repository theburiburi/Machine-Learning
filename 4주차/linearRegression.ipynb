```python
import sklearn
print(sklearn.__version__)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("LinearRegressionData.csv")
df.head()
```

```python
X = df.iloc[ : , :-1].values #왼쪽 = 행 , 오른쪽은 = 열 # 맨 뒤에서 앞까지
y = df.iloc[ : , -1].values

X, y
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() # 값 다 가져오기
lr.fit(X, y)
```

```python
y_pred = lr.predict(X)
y_pred
```

```python
plt.scatter(X, y, color = 'blue')
plt.plot(X, y_pred, color = 'red')
plt.title("score of hours")
plt.xlabel("hour")
plt.ylabel("score")
plt.show
```

```python
lr.predict([[11]])  #2차원으로 넣어야함, 예측해서 알려줌 # y hat 구한거임
```
